```python
# === Configuration ===
logs_path = "log_files/"
log_files_dict = {
                    '1': 'testing_wind_estimation_with17Ywind.log',
                }
```

## Get the data out

```python
# Reading the .log file into a pandas DataFrame
df_log = pd.read_csv(logs_path + log_files_dict['1'], header=0)
# Display the first few rows to confirm successful import
df_log.head()
```

```python
# calculate average and max ctrl comp time and rollout comptime
avg_ctrl_comp_time = df_log['ctrl_comp_time'].mean()
max_ctrl_comp_time = df_log['ctrl_comp_time'].max()
avg_rollout_comp_time = df_log['rollout_comptime'].mean()
max_rollout_comp_time = df_log['rollout_comptime'].max()
```

```python
print(f"Average Control Computation Time: {avg_ctrl_comp_time} ms")
print(f"Standard Deviation Control Computation Time: {df_log['ctrl_comp_time'].std()} ms")
print(f"Maximum Control Computation Time: {max_ctrl_comp_time} ms")


print(f"Average Rollout Computation Time: {avg_rollout_comp_time} ms")
print(f"Standard Deviation Rollout Computation Time: {df_log['rollout_comptime'].std()} ms")
print(f"Maximum Rollout Computation Time: {max_rollout_comp_time} ms")
```
